## Data Pre-processing
The goal is to separate stop words from non-stopwords 

Import G.K. Chesternson's The Man Who Was Thursday

```python
corpus = nltk.corpus.gutenberg.raw('chesterton-thursday.txt')
```

```python
tokens = nltk.tokenize.word_tokenize(corpus)
```

```python
new_tokens = [word.lower() for word in tokens if word.isalpha()]
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
words_excl_stopwords = [word for word in new_tokens if word not in stopwords]
```

```python
words_incl_stopwords = [word for word in new_tokens if word in stopwords]
```
